```python
import cv2
import os
from datetime import datetime
def capture_and_save_image(save_dir="captured_images"):
    if not os.path.exists(save_dir):
        os.makedirs(save_dir)

    cap = cv2.VideoCapture(1)
    if not cap.isOpened():
        print("Error: Could not open video capture.")
        return

    while True:
        ret, frame = cap.read()
        if not ret:
            print("Failed to grab frame.")
            break

        cv2.imshow("Press 'w' to capture, 'q' to quit", frame)

        key = cv2.waitKey(1) & 0xFF
        if key == ord('w'):
            timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
            filename = os.path.join(save_dir, f"capture_{timestamp}.jpg")
            cv2.imwrite(filename, frame)
            print(f"Image saved as {filename}")
        elif key == ord('q'):
            break

    cap.release()
    cv2.destroyAllWindows()

capture_and_save_image()
```

```python

import spacy
spacy.prefer_gpu()
nlp = spacy.load("sample_model1")
```

```python
list1 = []
dates_dic = {}
dictionary = {}
dates=[]
```

```python
from datetime import datetime

def parse_and_categorize_dates(dates):
    def parse_date(date_str):
        formats = ["%d %b %Y", "%b %Y", "%m/%y", "%d/%m/%y", "%d/%m/%Y"]
        for fmt in formats:
            try:
                return datetime.strptime(date_str, fmt)
            except ValueError:
                continue
        raise ValueError(f"Date format not recognized: {date_str}")

    categorized_dates = {"MANUFACTURER_DATE": [], "EXPIRY_DATE": []}
    for date in dates:
        date_obj = parse_date(date)
        if date_obj < datetime.now():
            categorized_dates["MANUFACTURER_DATE"].append(date)
        else:
            categorized_dates["EXPIRY_DATE"].append(date)

    return categorized_dates
```

```python
doc = nlp(extracted_text)
for ent in doc.ents:
    if ent.label_ == "EXPIRY_DATE":
        dates.append(ent.text)
    elif ent.label_ not in list1 and ent.text != "EXPIRY_DATE":
        dictionary[ent.label_] = ent.text
        list1.append(ent.label_)
dates
dates_dic = parse_and_categorize_dates(dates)
dictionary.update(dates_dic)
for i in dictionary:
    print(i,": ",dictionary[i])
```
